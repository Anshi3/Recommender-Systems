```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = {
    'title': [
        'Sacred Games', 'Mirzapur', 'Paatal Lok', 'The Family Man',
        'Delhi Crime', 'Made in Heaven', 'Kota Factory',
        'Scam 1992', 'Aspirants', 'Rocket Boys'
    ],
    'genre': [
        'Crime, Thriller', 'Crime, Action', 'Crime, Drama', 'Action, Drama',
        'Crime, Thriller', 'Drama, Romance', 'Drama, Comedy',
        'Biography, Crime', 'Drama', 'Biography, Drama'
    ]
}

df = pd.DataFrame(data)
```

```python
df.head()
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.metrics.pairwise  import cosine_similarity
```

```python
# Step 1: Convert genres into matrix of token counts

cv=CountVectorizer()
count_matrix=cv.fit_transform(df['genre'])
```

```python
# Step2 Cosine Similarity

similarity=cosine_similarity(count_matrix)
```

```python
# Step 3->Recommending similar Shows

def recommend(title):
    index=df[df['title']==title].index[0]
    distances=similarity[index]
    series_list=sorted(list(enumerate(distances)),reverse=True,key=lambda x:x[1])[1:4]

    print(f"\nðŸ’¡ Because you liked '{title}', you may also like:")
    for i in series_list:
            print(f"{df.iloc[i[0]]['title']}")

recommend('Kota Factory')
```
